````python
def updateList():
    """Refresh the displayed to-do list content"""
    
    #Clear the current list displayed on the screen   
    for widget in frame2.winfo_children():
        widget.destroy()
    
    #Generate and display the new list
    r = 0
    for task in todos:
        listFrame = Frame(frame2, relief = SUNKEN, height = 50)
        listFrame.pack(fill="x")
        r += 1
        taskLabel = Label(listFrame, text = "["+ str(r) + "]\t" + task, font="Helvetica 14")
        taskLabel.pack(side=LEFT)
        completeCheck = Checkbutton(listFrame, variable=completeInt, offvalue=0, onvalue=r, command=deleteTask)
        completeCheck.deselect()
        completeCheck.pack(side=RIGHT)
        
````
